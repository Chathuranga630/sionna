```python
import numpy as np
import matplotlib.pyplot as plt

# Base station positions
bs_pos = np.array([
    [0,0],
    [500,0]
])

# Random users
num_users = 20

user_pos = np.random.rand(num_users,2)*500

plt.scatter(user_pos[:,0],user_pos[:,1],label='Users')
plt.scatter(bs_pos[:,0],bs_pos[:,1],s=200,label='Base Stations')

plt.legend()
plt.grid()
plt.title("Micro-cell Network")
plt.savefig("network.png")
plt.show()
```

```python
import numpy as np

bs_pos = np.array([
    [0,0],
    [500,0]
])

num_users = 20

user_pos = np.random.rand(num_users,2)*500

assignments = []

for user in user_pos:

    d1 = np.linalg.norm(user-bs_pos[0])
    d2 = np.linalg.norm(user-bs_pos[1])

    if d1 < d2:
        assignments.append(0)
    else:
        assignments.append(1)

print(assignments)

print("Users connected to BS1:",assignments.count(0))
print("Users connected to BS2:",assignments.count(1))
```

```python
P_idle = 30
P_user = 2

normal_users = 10
attack_users = 40

normal_power = P_idle + P_user*normal_users
attack_power = P_idle + P_user*attack_users

print("Normal Power =",normal_power,"W")
print("Attack Power =",attack_power,"W")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulation area (meters)
AREA_X = 1000
AREA_Y = 1000

print("Simulation Area:", AREA_X, "x", AREA_Y, "meters")
```

```python
bs_positions = np.array([
    [250,250],
    [750,250],
    [500,750]
])

print(bs_positions)
```

```python
num_users = 30

user_positions = np.random.rand(num_users,2)

user_positions[:,0] *= AREA_X
user_positions[:,1] *= AREA_Y

print(user_positions)
```

```python
plt.figure(figsize=(8,8))

# Users
plt.scatter(
    user_positions[:,0],
    user_positions[:,1],
    c='blue',
    label='Users'
)

# Base Stations
plt.scatter(
    bs_positions[:,0],
    bs_positions[:,1],
    c='red',
    s=250,
    marker='^',
    label='Base Stations'
)

plt.xlim(0,AREA_X)
plt.ylim(0,AREA_Y)

plt.grid(True)
plt.legend()

plt.title("Micro-cell Network")

plt.show()
```
