```python
import numpy as np
import matplotlib.pyplot as plt

# Base station positions
bs_pos = np.array([
    [0,0],
    [500,0]
])

# Random users
num_users = 20

user_pos = np.random.rand(num_users,2)*500

plt.scatter(user_pos[:,0],user_pos[:,1],label='Users')
plt.scatter(bs_pos[:,0],bs_pos[:,1],s=200,label='Base Stations')

plt.legend()
plt.grid()
plt.title("Micro-cell Network")
plt.savefig("network.png")
plt.show()
```
