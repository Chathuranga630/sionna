```python
import numpy as np
import matplotlib.pyplot as plt

# Base station positions
bs_pos = np.array([
    [0,0],
    [500,0]
])

# Random users
num_users = 20

user_pos = np.random.rand(num_users,2)*500

plt.scatter(user_pos[:,0],user_pos[:,1],label='Users')
plt.scatter(bs_pos[:,0],bs_pos[:,1],s=200,label='Base Stations')

plt.legend()
plt.grid()
plt.title("Micro-cell Network")
plt.savefig("network.png")
plt.show()
```

```python
import numpy as np

bs_pos = np.array([
    [0,0],
    [500,0]
])

num_users = 20

user_pos = np.random.rand(num_users,2)*500

assignments = []

for user in user_pos:

    d1 = np.linalg.norm(user-bs_pos[0])
    d2 = np.linalg.norm(user-bs_pos[1])

    if d1 < d2:
        assignments.append(0)
    else:
        assignments.append(1)

print(assignments)

print("Users connected to BS1:",assignments.count(0))
print("Users connected to BS2:",assignments.count(1))
```

```python
capacity = 100

normal_load = 10
attack_load = 40

throughput_normal = capacity/normal_load
throughput_attack = capacity/attack_load

print("Normal:",throughput_normal)
print("Attack:",throughput_attack)
```

```python
P_idle = 30
P_user = 2

normal_users = 10
attack_users = 40

normal_power = P_idle + P_user*normal_users
attack_power = P_idle + P_user*attack_users

print("Normal Power =",normal_power,"W")
print("Attack Power =",attack_power,"W")
```
